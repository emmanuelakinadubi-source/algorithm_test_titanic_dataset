Now that the dataset is clean, we move on to regression analysis and modeling

All my import comes from the utils/import_util.py to keep the code clean

Since I am trying to predict the passengers that survived, my target variable is Surviced

```python
# Target
y = titanic_df['Survived']

# Features
X = titanic_df.drop(['Survived'], axis=1)
```

Columns not very useful
Name - too unique, Ticket - messy text, PassenderId - identifier

```python
# Drop unnecessary columns
X = X.drop(['Name', 'Ticket', 'PassengerId'], axis=1)
```

```python
X = pd.get_dummies(
    X,
    columns=X.select_dtypes(include=["object"]).columns.tolist(),
    drop_first=True
)
```

```python
# Converting the data to float
X = X.astype(float)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

# Linear regression is not the best model for this type of dataset
# WHY? Survived = 0 or 1 - Hence, Classification

```python
# A better model here is Logic Regression
# why? - Linear regression predicts continious values
# Logistic Regression predicts Categories/classes 
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
log_rig = LogisticRegression(max_iter=1000)
log_rig.fit(x_train, y_train)
```

```python
predictions = log_rig.predict(x_test)
```

```python


accuracy_score(y_test, predictions)
```

```python
y_pred_train = log_rig.predict(x_train)
```

```python
mean_squared_error(y_train, y_pred_train)
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = log_rig.predict(x_test)
print(mean_squared_error(y_test, y_pred_test))
print(r2_score(y_test, y_pred_test))
```
